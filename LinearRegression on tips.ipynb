```python
import seaborn as sns
df = sns.load_dataset('tips')
df.head()
```

```python
# one hot encoding
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder()
```

```python
import pandas as pd
col_encoder = encoder.fit_transform(df[['sex', 'smoker','day','time']]).toarray()
encoder_df = pd.DataFrame(col_encoder, columns=encoder.get_feature_names_out())
```

```python
df_copy = df.drop(columns = ['sex', 'smoker','day','time'], axis=1)
```

```python
final_df = pd.concat([df_copy, encoder_df], axis=1)
final_df
```

```python
final_df.info()
```

```python
## independent and dependent features
X = final_df.drop(columns=['total_bill'], axis=1)
y = final_df['total_bill']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.20, random_state=42)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mse)
print(mse, mae, rmse)
```

```python
print(r2_score(y_test, y_pred))
```
